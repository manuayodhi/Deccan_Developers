```python
import pandas as pd
import io
```

```python
file_path = 'dataset'
skip_rows = 0
with open(file_path, 'r') as f:
    for i, line in enumerate(f):
        if line.strip() == '@DATA':
            skip_rows = i + 1
            break
```

```python
df = pd.read_csv(file_path, skiprows=skip_rows, header=None)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,roc_auc_score
```

```python
rfc=RandomForestClassifier()
```

```python
df.columns
```

```python
df.drop(['surname','surname_tfidf_0','surname_tfidf_1','surname_tfidf_2','surname_tfidf_3', 'surname_tfidf_4'],axis=1,inplace=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop('exited',axis=1),df.exited,test_size=0.25,random_state=25)
```

```python
rfc=RandomForestClassifier(max_depth=4,
min_samples_leaf=4,
min_samples_split=2,
n_estimators=80)
```

```python
rfc.fit(x_train,y_train)
```

```python
y_pred1=rfc.predict(x_train)
y_pred2=rfc.predict(x_test)
```

```python
print(rfc.score(x_train,y_train))
print(rfc.score(x_test,y_test))
```

```python
# Classification Report for Training Data
print(classification_report(y_train, y_pred1))
```

```python
#Classification Report for Test Data
print(classification_report(y_test, y_pred2))
```
